```python
housing=fetch_california_housing()
m,n=housing.data.shape
housing_data_plus_bias=np.c_[np.ones((m,1)),housing.data]
X=tf.constant(housing_data_plus_bias,dtype=tf.float32,name="X")
y=tf.constant(housing.target.reshape(-1,1),dtype=tf.float32,name="y")
XT=tf.transpose(X)
theta=tf.matmul(tf.matmul(tf.linalg.inv(tf.matmul(XT,X)),XT),y)
print(theta.numpy())
```

```python
from sklearn.preprocessing import StandardScaler
housing_data_plus_bias=np.c_[np.ones((m,1)),housing.data]
#标准化训练集
scaler=StandardScaler()
scaled_housing_data_plus_bias=scaler.fit_transform(housing_data_plus_bias)
X=tf.constant(scaled_housing_data_plus_bias,dtype=tf.float32,name="X")
y=tf.constant(housing.target.reshape(-1,1),dtype=tf.float32)

#初始化参数
m,n=X.shape
theta=tf.Variable(tf.random.uniform(shape=[n,1],minval=-1.0,maxval=1.0),name="theta")

#定义模型
def linear_model(X,theta):
    return tf.matmul(X,theta)
    

#定义损失函数
def mean_square_error(y,y_pred):
    return tf.reduce_mean(tf.square(y,y_pred))

#自动计算梯度
def train_step(X,y,theta,learning_rate):
    y_pred=linear_model(X,theta)
    current_loss=mean_square_error(y,y_pred)
    error=y_pred-y
    current_loss=tf.reduce_mean(tf.square(error), name="mse")
    gradients=2/m*tf.matmul(tf.transpose(X),error)
    theta.assign_sub(theta-learning_rate*gradients)
    return current_loss

#训练循环
n_epochs=1000
learning_rate=0.01
for epoch in range(n_epochs):
    current_loss=train_step(X,y,theta,learning_rate)
    print("Epoch",epoch,"MSE=",current_loss.numpy())
best_theta=theta.numpy()
print(best_theta)
```

```python
#标准化训练集
scaler=StandardScaler()
scaled_housing_data_plus_bias=scaler.fit_transform(housing_data_plus_bias)
X=tf.constant(scaled_housing_data_plus_bias,dtype=tf.float32,name="X")
y=tf.constant(housing.target.reshape(-1,1),dtype=tf.float32)

#初始化参数
m,n=X.shape
theta=tf.Variable(tf.random.uniform(shape=[n,1],minval=-1.0,maxval=1.0),name="theta")
learning_rate=0.01
momentum=0.9

#创建梯度下降优化器
optimizer = tf.keras.optimizers.SGD(learning_rate=learning_rate,momentum=momentum)
#定义模型
model = tf.keras.Sequential([
  tf.keras.layers.Dense(1, input_shape=(n,))
])
# 编译模型，指定优化器和损失函数
model.compile(optimizer=optimizer, loss='mean_squared_error')
# 训练模型
model.fit(X, y, epochs=10)
```
