```python
import os
os.environ["OMP_NUM_THREADS"]="7"
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#导入数据
data=pd.read_csv(r".\mimic_perform_train_all_001_data.csv")
data.head()
```

```python
t=np.array(data["Time"])
y=np.array(data["PPG"])
# 进行傅里叶变换
fft_result = np.fft.fft(y)

# 获取频率轴的值
f = np.fft.fftfreq(y.size, d=t[1] - t[0])

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, y)
plt.title('Time Domain Signal')
plt.xlabel('Time (s)')
plt.ylabel('PPG')

plt.subplot(2, 1, 2)
plt.stem(f, np.abs(fft_result))  # 使用stem函数绘制频谱
plt.title('Frequency Domain Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
```
