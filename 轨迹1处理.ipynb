```python
import os
os.environ["OMP_NUM_THREADS"]="7"
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#导入数据
data=pd.read_csv(r".\轨迹1.csv")
x1=np.array(data["x1"])
y1=np.array(data["y1"])
x2=np.array(data["x2"])
y2=np.array(data["y2"])
x3=np.array(data["x3"])
y3=np.array(data["y3"])
```

```python
def square(x,y,data,colume_name):
    row_num=x.shape[0]
    result=[0]
    for i in range(row_num-1):
        distance=((x[i+1]-x[i])**2+(y[i+1]-y[i])**2)**0.5
        result.append(distance)
    data[colume_name]=np.array(result)
    return data

colume_name="distance_1"
square(x1,y1,data,colume_name)
colume_name="distance_2"
square(x2,y2,data,colume_name)
colume_name="distance_3"
square(x3,y3,data,colume_name)
data.to_csv(r"./轨迹1处理.csv",encoding='gbk')
data.head()
```

```python
def square_1(x,y,data,colume_name):
    row_num=x.shape[0]
    result1=[0,0]
    for i in range(0,row_num-2):
        distance1=((x[i+2]-x[i])**2+(y[i+2]-y[i])**2)**0.5
        result1.append(distance1)
    data[colume_name]=np.array(result1)
    return data

colume_name="distance_11"
square_1(x1,y1,data,colume_name)
colume_name="distance_22"
square_1(x2,y2,data,colume_name)
colume_name="distance_33"
square_1(x3,y3,data,colume_name)
data.to_csv(r"./轨迹1处理.csv",encoding='gbk')
data.head()
```
